```python
from langgraph.graph import StateGraph, START , END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatOpenAI()
```

```python
class BlogState(TypedDict):
    title : str
    outline : str
    content : str
    score : str
```

```python
def create_outline(state: BlogState)-> BlogState:
    title = state['title']

    prompt = f'Generate a detailed outline for a blog on the topic {title}'

    outline = model.invoke(prompt).content

    state['outline'] = outline

    return state
```

```python
def create_blog(state: BlogState)-> BlogState:
    title = state['title']
    outline = state['outline']

    prompt = f'Create a blog post on the title - {title} using the following outline \n {outline}.'

    content = model.invoke(prompt).content

    state['content'] = content

    return state
```

```python
def generate_score(state: BlogState) -> BlogState:
    title = state['title']
    outline = state['outline']
    content = state['content']

    prompt = f'Evaluate the following content of blog post \n {content} based on following outline \n {outline} and assign score out of 10 and return a number.'
    score = model.invoke(prompt).content

    state['score'] = score
    return state
    
```

```python
graph = StateGraph(BlogState)

# nodes
graph.add_node('create_outline', create_outline)
graph.add_node('create_blog', create_blog)
graph.add_node('generate_score', generate_score)

graph.add_edge(START, 'create_outline')
graph.add_edge('create_outline', 'create_blog')
graph.add_edge('create_blog', 'generate_score')
graph.add_edge('generate_score', END)

workflow = graph.compile()
```

```python
initial_state = {'title': 'Rise of AI in Italy'}

final_state = workflow.invoke(initial_state)
```

```python
print(f'score: {final_state['score']}')
print(final_state['content'])
```
